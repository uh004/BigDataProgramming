## Lab 2 (혼자해 보기) : mpg 데이터 분석

```python
mpg = pd.read_csv('mpg.csv')
mpg.head()
```

```python
mpg_a = mpg.query('displ < 4')
mpg_a.head()
```

```python
mpg_b = mpg.query('displ > 5')
mpg_b.head()
```

```python
mpg_a['hwy'].mean()
```

```python
mpg_b['hwy'].mean()
```

```python
if mpg_a['hwy'].mean() > mpg_b['hwy'].mean():
    print('배기량 4 이하 자동차의 고속도로 연비가 높다.')
elif mpg_a['hwy'].mean() < mpg_b['hwy'].mean():
    print('배기량 5 이상 자동차의 고속도로 연비가 높다')
```

```python
mpg_audi = mpg.query('manufacturer == "audi"')
mpg_audi.head()
```

```python
mpg_toyota = mpg.query('manufacturer == "toyota"')
mpg_toyota.head()
```

```python
mpg_audi['cty'].mean()
```

```python
mpg_toyota['cty'].mean()
```

```python
mpg_new = mpg.query('manufacturer in ["chevrolet","ford", "honda"]')
mpg_new['manufacturer'].value_counts()
```

```python
mpg_new['cty'].mean()
```

## Lab 7 : 파생변수 추가하기

```python
import pandas as pd
exam = pd.read_csv('exam.csv')
exam.head(3)
```

```python
exam.assign(total = exam['math'] + exam['english'] + exam['science'])
# exam.head()
```

```python
exam.assign(
        total = exam['math'] + exam['english'] +exam['science'],
        mean = (exam['math'] + exam['english'] + exam['science']) / 3)
# exam.head()
```

```python
import numpy as np
```

```python
exam.assign(test = np.where(exam['science'] > 60, 'pass', 'fail'))
# exam.head()
```

```python
exam.assign(total = exam['math'] + exam['english'] + exam['science']).sort_values('total')
```

## Lab 7 (혼자서 해보기) : mpg 데이터를 이용하여 문제 분석

```python
mpg = pd.read_csv('mpg.csv')
```

```python
mpg_new = mpg.copy()
```

```python
mpg_new = mpg_new.assign(total = mpg_new['cty'] + mpg_new['hwy'])
mpg_new.head(3)
```

```python
mpg_new = mpg_new.assign(mean = mpg_new['total']/2)
mpg_new.head(3)
```

```python
mpg_new.sort_values('mean', ascending=False).head(3)
```

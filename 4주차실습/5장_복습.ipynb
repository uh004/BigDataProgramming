# 5장 복습 하기 : mpg 데이터 파악해 보기

```python
import pandas as pd
mpg = pd.read_csv('mpg.csv')
mpg.head()
```

```python
mpg = mpg.rename(columns = {'hwy' : 'highway'})
mpg.head()
```

```python
mpg = mpg.rename(columns = {'cty' : 'city'})
mpg.head()
```

```python
mpg['total'] = mpg['city'] + mpg['highway']
mpg.head()
```

```python
mpg['mean'] = (mpg['city'] + mpg['highway']) / 2
mpg.head()
```

```python
mpg['total'].describe()
```

```python
import numpy as np
mpg['grade'] = np.where(mpg['total'] >= 47, 'A',
                np.where(mpg['total'] >= 31, 'B', 'C'))
mpg.head()
```

```python
count_grade = mpg['grade'].value_counts().sort_index()
count_grade
```

```python
count_grade.plot.bar(rot=0)
```

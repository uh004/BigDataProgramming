# Lab 2 : 가장 많이 사용된 단어 알아보기

```python
moon = open('speech_moon.txt', encoding = 'UTF-8').read()
moon
```

```python
import re
```

```python
moon = re.sub('[^가-힣]', ' ', moon)
moon
```

```python
import konlpy
hannanum = konlpy.tag.Hannanum()
```

```python
hannanum.pos('대한민국의 영토는 한반도와 그 부속도서로 한다') # N 명사 J 조사
```

```python
hannanum.tagset # 형태소 종류 알파벳 의미 확인
```

```python
hannanum.nouns('대한민국의 영토는 한반도와 그 부속도서로 한다') # 명사 추출하기
```

```python
hannanum.morphs('대한민국의 영토는 한반도와 그 부속도서로 한다') # 형태소 단위로 추출하기
```

```python
nouns = hannanum.nouns(moon) # 연설문에서 명사 추출하기
nouns[:20]
```

```python
import pandas as pd
df_word = pd.DataFrame({'word' : nouns}) # 데이터 프레임으로 변환
df_word
```

```python
df_word['word_n'] = df_word['word'].str.len() # 글자 수 파생 변수 추가
df_word
```

```python
df_word = df_word.query('word_n >= 2') # 단어 중 2글자 이상의 단어만 남기기
df_word.sort_values('word')
```

```python
df_word = df_word.groupby('word', as_index=False)\
    .agg(n = ('word', 'count'))\
    .sort_values('n', ascending=False)
df_word
```

```python
top20 = df_word.head(20)
top20
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.rcParams.update({'font.family' : 'malgun Gothic',
                     'figure.dpi' : '120',
                    'figure.figsize' : [6.5, 6]})

sns.barplot(data=top20, y='word', x='n')
```

# Lab 3 : 워드 클라우드 만들기

```python
!pip install wordcloud
```

```python
font = 'DoHyeon-Regular.ttf'
```

```python
df_word.head()
```

```python
dic_word = df_word.set_index('word').to_dict()['n']
dic_word
```

```python
from wordcloud import WordCloud
```

```python
wc = WordCloud(random_state = 1234,
              font_path = font,
              width = 400,
              height = 400,
              background_color = 'white')

img_wordcloud = wc.generate_from_frequencies(dic_word)

plt.figure(figsize=(5,5))
plt.axis('off')
plt.imshow(img_wordcloud)
```

# Lab 4 : 워드 클라우드 모양 바꾸기

```python
import PIL
```

```python
icon = PIL.Image.open('cloud.png')
```

```python
print(f'이미지 파일 이름 : {icon.filename}')
print(f'이미지 파일 형식 : {icon.format}')
print(f'이미지 파일 크기 : {icon.size}')
print(f'이미지 색상 모드 : {icon.mode}')
print(f'이미지 가로 크기 : {icon.width}')
print(f'이미지 세로 크기 : {icon.height}')
icon.show()
```

```python
import numpy as np

img = PIL.Image.new('RGB', icon.size, (255, 255, 255))

img.paste(icon, icon)

img = np.array(img)
```

```python
wc = WordCloud(random_state=1234,
              font_path = font,
              width =400,
              height= 400,
              contour_width=3,
              contour_color='steelblue',
              background_color = 'white',
              mask = img)

img_wordcloud = wc.generate_from_frequencies(dic_word)

plt.figure(figsize=(10,10))
plt.axis('off')
plt.imshow(img_wordcloud)
```

# Lab 6 : 가장 많이 사용된 단어 알아보기

```python
import pandas as pd
df = pd.read_csv('news_comment_BTS.csv', encoding='UTF-8')
```

```python
df.head()
```

```python
df.info()
```

```python
df['reply'] = df['reply'].str.replace('[^가-힣]', ' ', regex=True)
df['reply'].head()
```

```python
import konlpy
kkma = konlpy.tag.Kkma()
```

```python
nouns = df['reply'].apply(kkma.nouns) # 꼬꼬마 형태소 분석기로 명사 추출
nouns
```

```python
nouns = nouns.explode()
nouns
```

```python
type(nouns)
```

```python
df_word = pd.DataFrame({'word' : nouns})
df_word.head()
```

```python
df_word['count'] = df_word['word'].str.len()
df_word.head()
```

```python
df_word = df_word.query('count >= 2')
df_word.head()
```

```python
df_word = df_word.groupby('word', as_index=False)\
    .agg(n =('word', 'count'))\
    .sort_values('n', ascending=False)
df_word
```

```python
top20 = df_word.head(20)
top20
```

```python
plt.rcParams.update({'figure.figsize' : [6.5, 6]})

sns.barplot(data=top20, y='word', x='n')
```

```python
nouns[:30]
```

```python
df_word = df_word.query('word_n >= 2') # 단어 중 2글자 이상의 단어만 남기기
df_word
```

```python
df_word = df_word.groupby('word', as_index=False)\
    .agg(n = ('word', 'count'))\
    .sort_values('n', ascending=False)
df_word
```
